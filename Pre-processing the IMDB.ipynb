```python
import pandas as pd
# Importing Data
imdb2 = pd.read_csv('C:\\Users\\omers_000\\Dropbox\\DePaul\\CSC 478\\pika.csv')

# Dropping features with more than 40% missing values
imdb2 = imdb2.drop(["BoxOffice","DVD", "Website"],axis=1)
#Dropping features with same values
imdb2 = imdb2.drop(["Country","Language","Response","Type"],axis=1)
#Dropping categorical columns which can not be encoded or used
imdb2 = imdb2.drop(["Awards","Writer","imdbID"],axis=1)
#Dropping Metascore and imdbRatings because they are same as our dependent variable
imdb2 = imdb2.drop(["Metascore","imdbRatings"],axis=1)
#Dropping Released because it is same as Year
imdb2 = imdb2.drop(["Released"],axis=1)


#Dropping rows with missing values to create subset of Actors and Genres; and for binning Runtime
imdb2=imdb2.dropna()

# Dividing the Actors Column into 4 columns
Act_Sep = pd.DataFrame(imdb2.Actors.str.split(',',3).tolist(),
                                   columns = ['First Actor','Second Actor','Third Actor','Fourth Actor'])
Name = pd.DataFrame(imdb2["Name"].values)
Act_Sep = pd.concat([Act_Sep,Name], axis=1)
Act_Sep=Act_Sep.rename(columns = {0:'Name'})

#Merging four actors columns with original dataset
imdb2 = pd.merge(imdb2, Act_Sep, on="Name", how='inner')

# Dividing the Genre Column into 3 columns
Gen_Sep = pd.DataFrame(imdb2.Genre.str.split(',',3).tolist(),
                                   columns = ['First Genre','Second Genre','Third Genre'])
Gen_Sep = pd.concat([Gen_Sep,Name], axis=1)
Gen_Sep=Gen_Sep.rename(columns = {0:'Name'})

#Merging 3 genre columns with original dataset
imdb2 = pd.merge(imdb2, Gen_Sep, on="Name", how='inner')
imdb2=imdb2.drop(["Actors","Genre"],axis=1)

imdb2.Runtime = imdb2.Runtime.str.replace('min', '')
imdb2.Runtime = imdb2['Runtime'].astype(str).astype(int)

#Binning Runtime

run_bins = [0, 120, 566]
run_labels = ["Less than 2 hours","More than 2 hours"]
imdb2['Runtime'] = pd.cut(imdb2['Runtime'], run_bins, labels=run_labels, 
    right=True, include_lowest=True)

#Renaming
imdb2=imdb2.rename(columns = {"First Actor":'Actor1',"Second Actor":'Actor2',"Third Actor":'Actor3',"Fourth Actor":'Actor4',
                              "First Genre":'Genre1',"Second Genre":'Genre2',"Third Genre":'Genre3'})

imdb2
```
